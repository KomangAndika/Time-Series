## Autocorrelation
corellation measure linear relationship between two variables, autocorrelation measure linear relationship between lagged values of a time series. <br>


```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import numpy as np
import pandas as pd
np.random.seed(0)
time = pd.date_range(start='2023-01-01',periods=150,freq='D')
value = np.random.randn(150).cumsum()

df = pd.DataFrame({'time':time,'value':value})
df.set_index('time',inplace=True)
```

```python
plt.plot(df.index, df['value'])
```

```python
plt.figure(figsize=(10,6))
plot_acf(df['value'], lags=20)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('ACF graph')
plt.show
```

sometimes the plot is called correlogram.

## Seasonality and Trend in ACF Plots
When data have a trend, the autocorrelation for small lags tend to be larger because ot he uptrend or downtrend indicating high correlation.<br>
When the data are seasonal the autocorrelation will be larger for seasonal lags <br>
When it's a combination, the slow decrease of ACF as the lag increase is due to trend, while the shape is due to the seasonality.
